REMOVENDO COLUNAS

Para trabalhar com células vazias é bom remover linhas que tem valores vazios ou null. Isso é interessante quando se tem datasets enormes e que podem impactar no resultado de análise

```python
new_df = df.dropna()

print(new_df.to_string())
```

Observa-se que as linhas 18, 22 e 28 foram removidas por conter dados vazios

A função dropna() retorna um novo DataFrame e não muda o original. Se quiser mudar o original, use o comando inplace=True

```python
df.dropna(inplace = True)

print(df.to_string())

#Novo DataFrame ja modificado
```

PREENCHENDO OS VALORES VAZIOS

Outro jeito de lidar com os valores vazios, é preencher com outro dado no lugar. Use o comando fillna() para trocar os valores vazios por outros definidos.

No nosso exemplo, tem-se um valor vazio na coluna de Datas da linha 22, e valores vazios na coluna de calorias nas linhas 18 e 28. Então o ideal é preencher com valores específicos para cada uma das linhas

```python
import pandas as pd

df = pd.read_excel('C:/Users/lucas/Desktop/Cursos/Python/Pandas Python/BaseDeDados.xlsx')

df["Calories"].fillna(130, inplace = True)
df["Date"].fillna(2020/12/22, inplace=True)

#df["Nome da coluna que vai substituir"].fillna(valor substituido, inplace=True)
#Só substituir esse código pelos valores desejados a serem substituidos

print(df.to_string())
```

Repare que eu carreguei novamente o DF pois o comando inplace=true substituiu o df original. Substitui os valores NaN pelos desejados de acordo com as colunas escolhidas

Um bom jeito de preencher esses valores é usando a média, mediana ou moda dos valores da coluna

MODIFICANDO O FORMATO DOS DADOS

Repare que nas linhas 22 e 26 as datas estão incorretas, o ideal é padronizar os dados das datas. Use o comando datetime() para alterar esse problema


```python
import pandas as pd

df = pd.read_excel('C:/Users/lucas/Desktop/Cursos/Python/Pandas Python/BaseDeDados.xlsx')

df['Date'] = pd.to_datetime(df['Date'])

print(df.to_string())
```

A liha 26 foi arrumada, porém a 22 ainda não (NAT-Not A Time). Nesse caso é desejável remover a linha mesmo

```python
df.dropna(subset=['Date'], inplace = True)
print(df.to_string())
```

VALORES ERRADOS

Valores assim não precisam ser células vazias, ou em formatos errados, pode apenas estar literalmente com o valor errado. Exemplo: 199 ao invés de 1.99
